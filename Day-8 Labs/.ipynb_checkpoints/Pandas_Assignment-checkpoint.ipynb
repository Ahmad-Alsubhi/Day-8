# Q13: Rename all dataset's columns

|  Column name  |  New column name  |  Description  |
| ----- | -------- | ------- |
| pos    | Num_posts        
| flg    | Num_following    
| flr    | Num_followers    
| bl     | Biography_length 
| pic    | Picture_availability

| hc     | Average hashtag count 
| pr     | Promotional keywords
| fo     | Followers keywords 
| cs     | Cosine similarity  
| pi     | Post interval      
| class  | real_fake         

```python
dec={
    'pos': 'Num_posts',
    'flg': 'Num_following',
    'flr': 'Num_followers',
    'bl': 'Biography_length',
    'pic': 'Picture_availability',
    'lin': 'Link_availability',
    'cl': 'Average_caption_length',
    'cz': 'Caption_zero',
    'ni': 'Non_image_percentage',
    'erl': 'Engagement_rate_like',
    'erc': 'Engagement_rate_comment',
    'lt': 'Location_tag_percentage',
    'hc': 'Average_hashtag_count',
    'pr': 'Promotional_keywords',
    'fo': 'Followers_keywords',
    'cs': 'Cosine_similarity',
    'pi': 'Post_interval',
    'class': 'real_fake'   
}
edu.rename(columns=dec, inplace=True)
edu.head()

```

# Q14: Change the class's values to real and fake

```python
edu['real_fake'] = edu['real_fake'].map({'f': 'fake', 'r': 'real'})
print(edu['real_fake'].head())
```

# Q15: Print the total number of each fake accounts and real accounts 

```python
account_counts = edu['real_fake'].value_counts()
print("Number of fake accounts:", account_counts['fake'])
print("Number of real accounts:", account_counts['real'])
```
